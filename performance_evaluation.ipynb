# Evaluation of the results
* Varying epochs
* Varying learning rate
* Varying from 100 000 time steps to 1 000 000

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
def results_evaluation(table_path: str, x: str, x_label: str, plot_title: tuple[str, str]) -> None:
    """Plot the results and saves them

    Args:
        table_path (str): path to the results table
        x (str): x column
        x_label (str): label of the x column
        plot_title (tuple[str, str]): [policy, variable changed]
        image_saver (str): image name to be saved
    """
    # Import the results data
    results = pd.read_csv(table_path)

    # Plot the data
    results.plot(
        kind="line",
        x=x,
        y="total_profit",
        figsize=(15,6),
        ylim=(0, 2),
        ylabel="Total Profit",
        xlabel=x_label,
        title=f"{plot_title[0]} Total Profit Growth per Increase of {plot_title[1]}",
    )
    plt.legend(["Profit"])
    plt.grid(linestyle="--")

    # Save the image
    plt.savefig(
        fname=f"results/images_results/{plot_title[0]}_{plot_title[1]}.png", 
        dpi="figure", 
        format="png", 
        bbox_inches="tight",
        transparent=True
    )
```

```python
results_evaluation(
    table_path="results/tables/results.csv",
    x="num_epochs",
    x_label="Number of Epochs",
    plot_title=("MlpLstmPolicy", "Number of Epochs")
)
```

```python
results_evaluation(
    table_path="results/tables/results_lr.csv",
    x="learning_rate",
    x_label="Learning Rate",
    plot_title=("MlpLstmPolicy", "Learning Rate")
)
```
